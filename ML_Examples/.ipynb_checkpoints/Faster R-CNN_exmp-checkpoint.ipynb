## Faster R-CNN Algorithm example in PyTorch

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms.functional as F
```

```python
plt.rcParams["savefig.bbox"] = 'tight'


def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fig, axs = plt.subplots(ncols=len(imgs), squeeze=False)
    for i, img in enumerate(imgs):
        img = img.detach()
        img = F.to_pil_image(img)
        axs[0, i].imshow(np.asarray(img))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
```

```python
from torchvision.utils import make_grid
from torchvision.io import read_image
from pathlib import Path

forest_int = read_image("people_forest.jpg")
city_int = read_image("people_city.jpg")
img_list = [forest_int, city_int]
show(city_int)
show(forest_int)
```

```python
# Loading model
from torchvision.utils import draw_bounding_boxes
from torchvision.models.detection import fasterrcnn_resnet50_fpn, FasterRCNN_ResNet50_FPN_Weights

weights = FasterRCNN_ResNet50_FPN_Weights.DEFAULT
tfms = weights.transforms()
imgs = [tfms(i) for i in img_list]

model = fasterrcnn_resnet50_fpn(weights=weights, progress = False)
model = model.eval()

outputs = model(imgs)
print(outputs)
```

```python
scr_threshold = 0.95

imgs_with_bboxes = [
    draw_bounding_boxes(dog_int, boxes = output['boxes'][output['scores'] > scr_threshold], width = 3)
    for dog_int, output in zip(img_list, outputs)
]
show(imgs_with_bboxes[1])
```
