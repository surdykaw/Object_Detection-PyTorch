```python
import torch as tr
#import numpy as np
import matplotlib.pyplot as plt
```

```python
time = tr.arange(0,20).float()
speed = tr.randn(20)*3 + 0.75*(time-9.5)**2 + 1
plt.scatter(time, speed)
```

```python
def f(t, params):
    a,b,c = params
    return a*(t**2) + (b*t) + c

def mse(preds, targets):
    return ((preds-targets)**2).mean()
```

Etap 1: Inicjalizacja parametrów

```python
pms = tr.randn(3).requires_grad_()
pms
```

# Etap 2: Obliczanie prognoz (predykcje)

```python
preds = f(time, pms)
preds

def show_preds(preds, ax=None):
    if ax is None:
        ax = plt.subplots()[1]
        ax.scatter(time, speed)
        ax.scatter(time, tr.detach(preds).numpy(), color='red')
        ax.set_ylim(-300, 100)

show_preds(preds)
```

# Etap 3: Obliczanie straty

```python
loss = mse(preds, speed)
loss
```

# Etap 4: Obliczanie gradientów

```python
loss.backward()
pms.grad
pms
```

Etap 5: Stopniowanie wag

```python
lr = 1e-5
pms.data -= lr * pms.grad.data
pms.grad = None
```

```python
preds = f(time, pms)
mse(preds, speed)
show_preds(preds)
```

```python
# Funkcja zawrierajaca wszystkie w/w kroki
def apply_steps(params, prn=True):
    preds = f(time, params)
    loss = mse(preds, speed)
    loss.backward()
    params.data -= lr * params.grad.data
    params.grad = None
    if prn:
        print(loss.item())
    return preds
```

## Etap 6: Powtarzanie procesu

```python
for i in range(10):
    #apply_steps(pms)
    show_preds(apply_steps(pms, True))
```
